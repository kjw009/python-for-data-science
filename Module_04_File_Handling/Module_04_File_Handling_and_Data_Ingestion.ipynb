## 1. Reading and Writing Text Files

Think of a text file like a notebook on your desk. Python can **open** it, **read** what's inside, **write** new content, and **close** it when done.

There are two common modes:
- `"w"` — **write** (creates a new file, or overwrites an existing one)
- `"r"` — **read** (opens an existing file to read its contents)
- `"a"` — **append** (adds to the end of an existing file)


````python
# --- Writing to a text file ---
with open("my_notes.txt", "w") as f:
    f.write("Hello, this is line 1.\n")
    f.write("Python makes file handling easy!\n")
    f.write("This is line 3.\n")

print("✅ File written successfully!")
````

````python
# --- Reading the entire file at once ---
with open("my_notes.txt", "r") as f:
    content = f.read()

print(content)
````

````python
# --- Reading line by line (useful for large files) ---
with open("my_notes.txt", "r") as f:
    for line_number, line in enumerate(f, start=1):
        print(f"Line {line_number}: {line.strip()}")
````

**Key point:** Always use `with open(...)` — it automatically closes the file for you, even if an error occurs. This is called a **context manager**.

## 2. Working with CSV Files

CSV stands for **Comma-Separated Values**. It's one of the most common formats for storing tabular data (like a spreadsheet).

A CSV file looks like this:
```
name,age,city
Alice,30,London
Bob,25,Manchester
```


````python
# --- Create a sample CSV file ---
data = [
    ["name", "age", "city"],
    ["Alice", 30, "London"],
    ["Bob", 25, "Manchester"],
    ["Charlie", 35, "Edinburgh"]
]

with open("people.csv", "w", newline="") as f:
    writer = csv.writer(f)
    writer.writerows(data)

print("✅ CSV file created!")
````

````python
# --- Read the CSV file back ---
with open("people.csv", "r") as f:
    reader = csv.reader(f)
    for row in reader:
        print(row)
````

````python
# --- Read CSV as a list of dictionaries (much easier to work with!) ---
with open("people.csv", "r") as f:
    reader = csv.DictReader(f)
    for row in reader:
        print(f"{row['name']} is {row['age']} years old, lives in {row['city']}")
````

**Why DictReader?** With `csv.reader`, you get plain lists like `['Alice', '30', 'London']`. With `csv.DictReader`, you get dictionaries like `{'name': 'Alice', 'age': '30', 'city': 'London'}` — much easier to understand and use.

## 3. Working with JSON Data

JSON (JavaScript Object Notation) is the most popular format for exchanging data on the web. It looks very similar to Python dictionaries!

```json
{"name": "Alice", "age": 30, "hobbies": ["reading", "coding"]}
```


````python
# --- Create and save a JSON file ---
student = {
    "name": "Alice",
    "age": 30,
    "courses": ["Python", "Data Science"],
    "graduated": False
}

with open("student.json", "w") as f:
    json.dump(student, f, indent=2)   # indent=2 makes it human-readable

print("✅ JSON file saved!")
````

````python
# --- Load JSON back into Python ---
with open("student.json", "r") as f:
    loaded = json.load(f)

print(type(loaded))          # It's a dictionary!
print(loaded["name"])        # Access like any dict
print(loaded["courses"][0])  # Access nested data
````

````python
info_box(
    "json.dump vs json.dumps",
    "Use <code>json.dump(data, file)</code> to write to a <b>file</b>.<br>"
    "Use <code>json.dumps(data)</code> to convert to a <b>string</b> (the 's' stands for 'string')."
)
````

## 4. Reading Excel Files with openpyxl

Many businesses store data in Excel spreadsheets (`.xlsx`). The `openpyxl` library lets Python read and write these files.


````python
# --- Create a simple Excel file ---
wb = openpyxl.Workbook()
ws = wb.active
ws.title = "Sales"

# Add header row
ws.append(["Product", "Quantity", "Price"])

# Add data rows
ws.append(["Laptop", 10, 999.99])
ws.append(["Mouse", 50, 25.50])
ws.append(["Keyboard", 30, 75.00])

wb.save("sales.xlsx")
print("✅ Excel file created!")
````

````python
# --- Read the Excel file ---
wb = openpyxl.load_workbook("sales.xlsx")
ws = wb.active

print(f"Sheet name: {ws.title}")
print(f"Number of rows: {ws.max_row}")
print()

for row in ws.iter_rows(values_only=True):
    print(row)
````

## 5. Web Scraping Basics

Web scraping means extracting data from websites. We use two libraries:
- `requests` — downloads the web page
- `BeautifulSoup` — parses the HTML so we can find specific content

Let's scrape a simple, beginner-friendly example.


````python
# --- Fetch a web page ---
url = "https://example.com"
response = requests.get(url)

print(f"Status code: {response.status_code}")  # 200 = success
print(f"Content length: {len(response.text)} characters")
````

````python
# --- Parse the HTML with BeautifulSoup ---
soup = BeautifulSoup(response.text, "html.parser")

# Get the page title
print("Page title:", soup.title.string)

# Get all paragraph text
for p in soup.find_all("p"):
    print("Paragraph:", p.get_text()[:80], "...")
````

````python
warning_box(
    "Be Respectful When Scraping",
    "Always check a website's <code>robots.txt</code> file and Terms of Service before scraping. "
    "Don't send too many requests too quickly — add pauses between requests with <code>time.sleep()</code>."
)
````

## 6. Working with APIs

An **API** (Application Programming Interface) is like a waiter in a restaurant — you tell it what you want (a request), and it brings you the data (a response). Most APIs return data in **JSON** format.

Let's use a free, public API to fetch some data.


````python
# --- Fetch data from a free public API ---
url = "https://api.github.com/users/python"
response = requests.get(url)

if response.status_code == 200:
    data = response.json()   # Parse JSON response into a dictionary
    print(f"Name: {data['name']}")
    print(f"Bio: {data['bio']}")
    print(f"Public repos: {data['public_repos']}")
    print(f"Followers: {data['followers']}")
else:
    print(f"Error: Status code {response.status_code}")
````

````python
info_box(
    "Common HTTP Status Codes",
    "<b>200</b> = OK (success)<br>"
    "<b>404</b> = Not Found (wrong URL)<br>"
    "<b>403</b> = Forbidden (no permission)<br>"
    "<b>500</b> = Server Error (problem on their side)"
)
````

## 🎯 Key Takeaways

1. **Text files** — Use `open()` with `with` statement. Modes: `"r"` (read), `"w"` (write), `"a"` (append).
2. **CSV files** — Use `csv.reader` for lists, `csv.DictReader` for dictionaries (much friendlier).
3. **JSON files** — Use `json.dump()` to save, `json.load()` to read. JSON ↔ Python dict is nearly 1-to-1.
4. **Excel files** — Use `openpyxl` to create and read `.xlsx` files.
5. **Web scraping** — `requests.get()` fetches pages; `BeautifulSoup` parses the HTML.
6. **APIs** — Send a request, get JSON back. Use `response.json()` to parse it.
