## 📖 Tuples — Unchangeable Lists

A **tuple** is like a list, but **you cannot modify it** after creation. It's **immutable**.

> **Analogy:** A list is a whiteboard (you can erase and rewrite). A tuple is a printed certificate (once printed, it's permanent).

### When to Use Tuples in Data Science

- **Coordinates:** `(latitude, longitude)` — these shouldn't change
- **Database rows:** Each row returned from a query is often a tuple
- **Function returns:** Functions that return multiple values return them as tuples
- **Dictionary keys:** Tuples can be dict keys (lists cannot)

````python
# Creating tuples — use parentheses ()
coordinates = (28.6139, 77.2090)   # Delhi lat/long
rgb_colour = (255, 128, 0)         # Orange
person = ("Alice", 25, "Mumbai")

print("Coordinates:", coordinates)
print("First value:", coordinates[0])   # Indexing works!
print("Slicing:", person[:2])           # Slicing works!
````

````python
# Tuples are IMMUTABLE — you can't change them
point = (10, 20)
# point[0] = 99   # ❌ TypeError! Can't modify a tuple

# This is actually useful — it PROTECTS your data from accidental changes
print("Point is safe:", point)
````

````python
# Tuple unpacking — a very handy Python feature!
# Instead of accessing by index, unpack into named variables

student = ("Priya", 24, "Data Science")
name, age, course = student   # Unpack into 3 variables

print(f"{name} is {age} years old, studying {course}")
````

````python
# Functions that return multiple values use tuples
def min_max(numbers):
    return min(numbers), max(numbers)   # Returns a tuple

data = [45, 12, 89, 34, 67]
lowest, highest = min_max(data)   # Unpack the result
print(f"Range: {lowest} to {highest}")
````

````python
info_box("When to Use Tuple vs List",
    "<b>Use a tuple</b> when the data is fixed and shouldn't change "
    "(coordinates, RGB colours, database records).<br>"
    "<b>Use a list</b> when you need to add, remove, or modify items "
    "(collecting results, building datasets).<br><br>"
    "Rule of thumb: if it's a <em>collection of different things</em> about one entity "
    "(name, age, city) → tuple. If it's a <em>collection of similar things</em> "
    "(list of scores) → list.")
````

## 📖 Sets — Unique Values Only

A **set** is an unordered collection with **no duplicates**. If you add a duplicate, it's ignored.

> **Analogy:** A set is like a **guest list** — each name appears only once, no matter how many times someone RSVPs.

### When to Use Sets in Data Science

- **Finding unique values** in a column
- **Removing duplicates** from data
- **Comparing groups** — what's in group A but not B?

````python
# Creating a set — use curly braces {} (same as dict, but no key:value)
fruits = {"apple", "banana", "cherry", "apple", "banana"}
print("Set:", fruits)       # Duplicates removed automatically!
print("Count:", len(fruits))  # 3, not 5
````

````python
# The #1 use case: finding unique values in data
customer_cities = ["Mumbai", "Delhi", "Mumbai", "Chennai", "Delhi", 
                   "Mumbai", "Bangalore", "Chennai", "Delhi"]

unique_cities = set(customer_cities)
print("All cities:", customer_cities)
print("Unique cities:", unique_cities)
print("Number of unique cities:", len(unique_cities))
````

````python
# Remove duplicates from a list (preserving → list conversion)
raw_data = [5, 3, 8, 3, 5, 1, 8, 9, 1]
unique_data = list(set(raw_data))
print("Deduplicated:", unique_data)
````

### Set Operations — Comparing Groups

This is where sets really shine in Data Science:

````python
# Imagine two marketing campaigns
campaign_a = {"Alice", "Bob", "Charlie", "Diana", "Eve"}
campaign_b = {"Charlie", "Diana", "Frank", "Grace"}

# Who was in BOTH campaigns? (intersection)
both = campaign_a & campaign_b
print("In both:", both)

# Who was in EITHER campaign? (union)
all_reached = campaign_a | campaign_b
print("All reached:", all_reached)

# Who was ONLY in campaign A? (difference)
only_a = campaign_a - campaign_b
print("Only in A:", only_a)

# Who was in one but NOT both? (symmetric difference)
exclusive = campaign_a ^ campaign_b
print("In one but not both:", exclusive)
````

````python
# Quick membership check — sets are MUCH faster than lists for this
large_vip_set = {"customer_" + str(i) for i in range(100000)}

# This is nearly instant with a set:
print("Is customer_99999 a VIP?", "customer_99999" in large_vip_set)
````

## 📖 Choosing the Right Data Structure

| Feature | List `[]` | Tuple `()` | Dictionary `{}` | Set `{}` |
|---------|-----------|------------|-----------------|----------|
| **Ordered?** | ✅ Yes | ✅ Yes | ✅ Yes (3.7+) | ❌ No |
| **Mutable?** | ✅ Yes | ❌ No | ✅ Yes | ✅ Yes |
| **Duplicates?** | ✅ Allowed | ✅ Allowed | ❌ Keys unique | ❌ No |
| **Access by** | Index `[0]` | Index `[0]` | Key `["name"]` | N/A |
| **Best for** | Sequences of similar items | Fixed records, returns | Labelled data, lookups | Unique values, comparisons |

### Quick Decision Guide

- **"I need an ordered collection I can modify"** → **List**
- **"I need to look up values by name"** → **Dictionary**
- **"I need unique values or to compare groups"** → **Set**
- **"I need a fixed, unchangeable record"** → **Tuple**

## 🎯 Key Takeaways

1. **Tuples** are immutable lists — use for fixed data, function returns, and coordinates
2. **Sets** hold unique values only — perfect for deduplication and group comparisons
3. Set operations (`&`, `|`, `-`) are powerful tools for comparing datasets
4. **Choose your structure based on the task**, not habit — the right choice makes your code simpler and faster
