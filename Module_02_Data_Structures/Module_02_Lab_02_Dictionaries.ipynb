## 📖 What is a Dictionary?

A **dictionary** stores data as **key-value pairs**. Instead of accessing items by position (like lists), you access them by a meaningful **name** (the key).

> **Analogy:** A dictionary is like a real dictionary — you look up a **word** (key) to find its **definition** (value). Or think of it as a contact card: the field name ("Phone") is the key, and the actual number is the value.

### Why Dictionaries Matter in Data Science

- JSON data from APIs? That's dictionaries.
- A single row in a dataset? Often represented as a dictionary.
- Configuration settings? Dictionary.
- Counting occurrences of categories? Dictionary.

````python
# Creating a dictionary — use curly braces {}
student = {
    "name": "Priya",
    "age": 24,
    "course": "Data Science",
    "gpa": 3.85
}

print(student)
````

````python
# Accessing values — use the KEY, not an index number
print("Name:", student["name"])
print("GPA:", student["gpa"])

# Safer access with .get() — returns None instead of crashing
print("Email:", student.get("email"))            # None (key doesn't exist)
print("Email:", student.get("email", "N/A"))     # "N/A" (custom default)
````

````python
warning_box("Use .get() for Safety",
    '<code>student["email"]</code> → crashes with KeyError if key doesn't exist<br>'
    '<code>student.get("email")</code> → safely returns None<br><br>'
    "In Data Science, data is often messy and keys may be missing. "
    "Always use <code>.get()</code> when you're not 100% sure the key exists.")
````

## 📖 Adding, Updating, and Removing Data

````python
student = {"name": "Priya", "age": 24}

# Add a new key-value pair
student["city"] = "Mumbai"
print("After adding city:", student)

# Update an existing value
student["age"] = 25
print("After birthday:", student)

# Remove a key
del student["city"]
print("After removing city:", student)

# .pop() removes AND returns the value
age = student.pop("age")
print("Popped age:", age)
print("Remaining:", student)
````

## 📖 Looping Through Dictionaries

````python
student = {"name": "Priya", "age": 24, "course": "Data Science", "gpa": 3.85}

# Loop through keys
print("Keys:")
for key in student:
    print(f"  {key}")
````

````python
# Loop through keys AND values (most common)
print("All data:")
for key, value in student.items():
    print(f"  {key}: {value}")
````

````python
# Loop through just the values
print("Values only:")
for value in student.values():
    print(f"  {value}")
````

## 📖 Data Science Patterns with Dictionaries

````python
# Pattern 1: Counting occurrences — very common in DS!
fruits = ["apple", "banana", "apple", "cherry", "banana", "apple", "date"]

count = {}
for fruit in fruits:
    if fruit in count:
        count[fruit] += 1
    else:
        count[fruit] = 1

print("Fruit counts:", count)
print("Most common:", max(count, key=count.get))
````

````python
# Pattern 2: A list of dictionaries = a dataset!
# Each dictionary is a row, each key is a column name.

students = [
    {"name": "Alice", "score": 92, "grade": "A"},
    {"name": "Bob", "score": 78, "grade": "C"},
    {"name": "Charlie", "score": 85, "grade": "B"},
    {"name": "Diana", "score": 95, "grade": "A"},
]

# This is basically how Pandas DataFrames work under the hood!
print("Student records:")
for s in students:
    print(f"  {s['name']}: {s['score']} ({s['grade']})")

# Extract just the scores
scores = [s["score"] for s in students]
print(f"\nAverage score: {sum(scores) / len(scores)}")
````

````python
info_box("Lists of Dictionaries → DataFrames",
    "The pattern above — a list of dictionaries where each dict has the same keys — "
    "is exactly how you'll create Pandas DataFrames later!<br><br>"
    "<code>import pandas as pd</code><br>"
    "<code>df = pd.DataFrame(students)</code><br>"
    "That one line turns our list of dicts into a full data table.")
````

````python
# Pattern 3: Dictionary comprehension (one-liner)
names = ["alice", "bob", "charlie"]
scores = [85, 92, 78]

# Create a dict from two lists using zip()
grade_book = {name: score for name, score in zip(names, scores)}
print("Grade book:", grade_book)
````

## 📖 Nested Dictionaries

Dictionaries can contain other dictionaries — this is how complex data like JSON is structured.

````python
# A nested dictionary — like a JSON response from an API
user = {
    "name": "Priya Sharma",
    "age": 24,
    "address": {
        "city": "Mumbai",
        "state": "Maharashtra",
        "pincode": "400001"
    },
    "scores": [85, 92, 78, 95]
}

# Accessing nested data
print("City:", user["address"]["city"])
print("First score:", user["scores"][0])
print("Average score:", sum(user["scores"]) / len(user["scores"]))
````
