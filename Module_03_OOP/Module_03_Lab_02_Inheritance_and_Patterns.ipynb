## 📖 Inheritance — Building on Existing Classes

**Inheritance** lets you create a new class based on an existing one. The new class gets all the data and methods of the original, and can add or change things.

> **Analogy:** Think of a **car**. A "Car" is the base class. A "Tesla" inherits everything a car has (wheels, engine, steering) but adds something new (electric motor, autopilot). You don't redesign the whole car — you build on top of it.

### Why This Matters in Data Science

- scikit-learn uses this everywhere: `LinearRegression`, `RandomForest` — they all inherit from a base `Estimator` class
- You can extend library classes to add your own custom behaviour

````python
# Base class (parent)
class Employee:
    def __init__(self, name, salary):
        self.name = name
        self.salary = salary
    
    def info(self):
        return f"{self.name}, Salary: {self.salary:,}"
    
    def __repr__(self):
        return f"Employee('{self.name}')"

# Child class — inherits everything from Employee
class DataScientist(Employee):
    def __init__(self, name, salary, languages):
        super().__init__(name, salary)   # Call parent's __init__
        self.languages = languages       # Add new attribute
    
    def info(self):
        base = super().info()            # Get parent's info
        langs = ", ".join(self.languages)
        return f"{base}, Languages: {langs}"

# Create objects
emp = Employee("Alice", 50000)
ds = DataScientist("Priya", 85000, ["Python", "SQL", "R"])

print(emp.info())
print(ds.info())     # Has everything Employee has, PLUS languages
````

````python
info_box("super() Explained",
    "<code>super().__init__(name, salary)</code> calls the <b>parent class's</b> __init__.<br>"
    "This means: do everything Employee.__init__ does first, then add my extra stuff.<br><br>"
    "Without <code>super()</code>, you'd have to rewrite all the parent's setup code.")
````

````python
# The child IS also an instance of the parent
print(f"Is ds an Employee? {isinstance(ds, Employee)}")         # True!
print(f"Is ds a DataScientist? {isinstance(ds, DataScientist)}") # True
print(f"Is emp a DataScientist? {isinstance(emp, DataScientist)}") # False
````

## 📖 Method Overriding — Customising Behaviour

When a child class defines a method with the **same name** as the parent, it **overrides** (replaces) it.

````python
class Shape:
    def area(self):
        return 0    # Default — no area
    
    def describe(self):
        name = type(self).__name__
        return f"{name}: area = {self.area():.2f}"

class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height
    
    def area(self):     # Override parent's area()
        return self.width * self.height

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius
    
    def area(self):     # Override parent's area()
        return 3.14159 * self.radius ** 2

# Both use the SAME describe() method from Shape,
# but each has its OWN area() calculation
shapes = [Rectangle(5, 3), Circle(7), Rectangle(10, 2)]

for shape in shapes:
    print(shape.describe())
````

````python
info_box("This is Polymorphism",
    "We called <code>.describe()</code> on different types of shapes, and each one "
    "calculated its area differently. The same method name, different behaviour — "
    "that's <b>polymorphism</b>.<br><br>"
    "scikit-learn does exactly this: <code>model.fit()</code> works on LinearRegression, "
    "RandomForest, SVM — same method name, different algorithms underneath.")
````

## 📖 `@property` — Computed Attributes

Sometimes you want an attribute that's **calculated on the fly** rather than stored. The `@property` decorator lets you access a method as if it were a regular attribute (no parentheses needed).

````python
class Student:
    def __init__(self, name, scores):
        self.name = name
        self.scores = scores
    
    @property
    def average(self):
        # Calculated automatically — always up to date
        return sum(self.scores) / len(self.scores)
    
    @property
    def grade(self):
        avg = self.average
        if avg >= 90: return "A"
        elif avg >= 80: return "B"
        elif avg >= 70: return "C"
        elif avg >= 60: return "D"
        else: return "F"
    
    def __repr__(self):
        return f"Student('{self.name}', avg={self.average:.1f}, grade={self.grade})"

priya = Student("Priya", [85, 92, 78, 95])

# Access like attributes — no () needed!
print(f"Average: {priya.average}")
print(f"Grade: {priya.grade}")
print(priya)

# If we add a new score, average and grade update automatically
priya.scores.append(60)
print(f"After new score: {priya}")
````

## 🚀 Practical Example: A Reusable Data Cleaner

Let's build something you'd actually use — a class that cleans messy data.

````python
class DataCleaner:
    # A reusable tool for cleaning tabular data.
    
    def __init__(self, data):
        self.data = [row.copy() for row in data]   # Don't modify original
        self.log = []    # Track what we did
    
    @property
    def num_rows(self):
        return len(self.data)
    
    def remove_duplicates(self):
        seen = []
        unique = []
        for row in self.data:
            row_str = str(sorted(row.items()))
            if row_str not in seen:
                seen.append(row_str)
                unique.append(row)
        removed = self.num_rows - len(unique)
        self.data = unique
        self.log.append(f"Removed {removed} duplicate rows")
        return self    # Return self for method chaining
    
    def fill_missing(self, column, default_value):
        count = 0
        for row in self.data:
            if row.get(column) is None or row.get(column) == "":
                row[column] = default_value
                count += 1
        self.log.append(f"Filled {count} missing values in '{column}'")
        return self
    
    def strip_strings(self):
        for row in self.data:
            for key, value in row.items():
                if isinstance(value, str):
                    row[key] = value.strip()
        self.log.append("Stripped whitespace from all strings")
        return self
    
    def report(self):
        print(f"Cleaning Report ({self.num_rows} rows remaining):")
        for step in self.log:
            print(f"  - {step}")
    
    def head(self, n=3):
        for row in self.data[:n]:
            print(row)

# Messy data
raw_data = [
    {"name": "  Alice ", "age": 25, "city": "Mumbai"},
    {"name": "Bob", "age": None, "city": "  Delhi  "},
    {"name": "Alice", "age": 25, "city": "Mumbai"},      # Duplicate
    {"name": " Charlie ", "age": 30, "city": ""},
    {"name": "Alice", "age": 25, "city": "Mumbai"},      # Duplicate
]

# Clean it — method chaining!
cleaner = (DataCleaner(raw_data)
           .remove_duplicates()
           .strip_strings()
           .fill_missing("city", "Unknown")
           .fill_missing("age", 0))

cleaner.report()
print()
cleaner.head(5)
````

````python
info_box("Method Chaining",
    "Notice how we wrote <code>.remove_duplicates().strip_strings().fill_missing(...)</code> "
    "all in one chain? Each method returns <code>self</code>.<br><br>"
    "Pandas uses this exact pattern:<br>"
    "<code>df.dropna().reset_index().sort_values('date')</code>")
````

class Manager(Employee):

        super().__init__(name, salary)

```python
class StatsSummary:
    def __init__(self, numbers):
        self.numbers = sorted(numbers)
    
    @property
    def mean(self):
        return sum(self.numbers) / len(self.numbers)
    
    @property
    def median(self):
        n = len(self.numbers)
        mid = n // 2
        if n % 2 == 0:
            return (self.numbers[mid - 1] + self.numbers[mid]) / 2
        return self.numbers[mid]
    
    @property
    def minimum(self):
        return self.numbers[0]
    
    @property
    def maximum(self):
        return self.numbers[-1]
    
    @property
    def range(self):
        return self.maximum - self.minimum
    
    def summary(self):
        print(f"Mean: {self.mean:.1f}")
        print(f"Median: {self.median}")
        print(f"Min: {self.minimum}")
        print(f"Max: {self.maximum}")
        print(f"Range: {self.range}")
```
